## 4. Preprocessing & Merge

Dataset yang digunakan untuk dataset akhir adalah **Sleep** dan **Stress**,
karena keduanya memiliki kolom yang relevan: `age`, `sex`, `sleep`, dan `stress`.

Kolom `fatigue` dibuat dengan logika sederhana:
- **Tinggi (2)** → tidur < 5 jam **DAN** stres > 6
- **Sedang (1)** → tidur < 6.5 jam **ATAU** stres > 5
- **Rendah (0)** → selain kondisi di atas

```python
import numpy as np

# --- Ekstrak dari dataset Sleep ---
df_sleep = sleep[['person_id', 'age', 'gender', 'sleep_duration_hrs', 'stress_score']].copy()
df_sleep.columns = ['id', 'age', 'sex', 'sleep', 'stress']
df_sleep['sex']   = df_sleep['sex'].map({'Female': 0, 'Male': 1})
df_sleep['shift'] = 1  # default shift reguler

# --- Ekstrak dari dataset Stress ---
df_stress = stress[['Employee_ID', 'Age', 'Gender', 'Sleep_Hours', 'Stress_Level']].copy()
df_stress.columns = ['id', 'age', 'sex', 'sleep', 'stress']
df_stress['sex']  = df_stress['sex'].map({'Female': 0, 'Male': 1})
df_stress['id']   = df_stress['id'].str.replace('E', '').astype(int)
df_stress['shift'] = 1

# --- Gabungkan kedua dataset secara vertikal ---
df_merged = pd.concat([df_sleep, df_stress], ignore_index=True)

# --- Buat kolom fatigue berdasarkan sleep & stress ---
def hitung_fatigue(row):
    if row['sleep'] < 5.0 and row['stress'] > 6.0:
        return 2  # Fatigue Tinggi
    elif row['sleep'] < 6.5 or row['stress'] > 5.0:
        return 1  # Fatigue Sedang
    else:
        return 0  # Fatigue Rendah

df_merged['fatigue'] = df_merged.apply(hitung_fatigue, axis=1)

# --- Susun kolom akhir ---
df_final = df_merged[['id', 'age', 'sex', 'sleep', 'shift', 'stress', 'fatigue']]

print(f"✅ Dataset final: {df_final.shape[0]:,} baris x {df_final.shape[1]} kolom")
print(f"   Distribusi fatigue: {df_final['fatigue'].value_counts().to_dict()}")
display(df_final.head())
```

### 5a. Ekstraksi Data Teks (NLP)
Kita akan mengambil kolom teks dari dataset Industrial Safety dan OSHA untuk membuat dataset NLP khusus.

```python
import pandas as pd

# Memastikan nama kolom yang benar dari OSHA (seringkali ada spasi atau perbedaan case)
osha_cols = osha.columns.tolist()
abstract_col = [c for c in osha_cols if 'ABSTRACT' in c.upper()][0]
print(f"Menggunakan kolom OSHA: {abstract_col}")

# 1. Ambil deskripsi dari Industrial Safety
df_nlp_industrial = industrial[['Description']].copy()
df_nlp_industrial.columns = ['text_description']
df_nlp_industrial['source'] = 'industrial_safety'

# 2. Ambil abstrak dari OSHA menggunakan nama kolom yang ditemukan
df_nlp_osha = osha[[abstract_col]].copy()
df_nlp_osha.columns = ['text_description']
df_nlp_osha['source'] = 'osha'

# 3. Gabungkan
df_nlp = pd.concat([df_nlp_industrial, df_nlp_osha], ignore_index=True)

# Bersihkan baris yang kosong jika ada
df_nlp = df_nlp.dropna(subset=['text_description'])

# Simpan ke nlp.csv
df_nlp.to_csv('/content/nlp.csv', index=False)

print(f"✅ Berhasil membuat nlp.csv dengan {len(df_nlp):,} baris.")
display(df_nlp.head())
```
